```python
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months, savings)# x variables are in first position, y variables are in second 
plt.xlabel('Month') #xlabel takes a single argument, a string that we want the x labels to represent
plt.ylabel('Savings ($)')
plt.title('Saving Account Results')
plt.show() # show should always be the final line of code
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() -  the object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Edit Formatting
#Change the line color to green.
#Change the marker to a diamond.
#Change the line width to 2 points.
#Add a legend for the city of Boston.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
plt.legend()

#Add a title and axes labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Add a y-axis limit.
plt.ylim(0, 45)
#Add grid lines
plt.grid()
```

```python
# Skill Drill
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)

#Change the bar color to cyan.
plt.barh(x_axis, y_axis, color="cyan", label='Chicago') 

#Add a title and axes labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Switch the axis so that the Fare($) data is on the x-axis.
plt.xlabel("Fare($)")
plt.ylabel("Date")

#Invert the y-axis data.
plt.gca().invert_yaxis()

#Add a legend for the city of Chicago.
plt.legend()
```
